```python
# Crear un nuevo DataFrame con las separaciones
def generar_df_separaciones(df):
    rows = []
    for _, row in df.iterrows():
        for st_id, texto_individual in enumerate(row['punt'], start=1):
            rows.append({
                'id': row['id'],  # ID del episodio original
                'st_id': st_id,   # ID individual de la separación
                'sentence': texto_individual  # Contenido de la separación
            })
    return pd.DataFrame(rows)
```

```python
df_separaciones = generar_df_separaciones(df)

df_separaciones
```

```python
import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
```

```python
# 1. Tokenización: Convertir frases en listas de palabras
df_separaciones['tokens'] = df_separaciones['sentence'].apply(lambda x: word_tokenize(x.lower()))

# 2. Entrenamiento de Word2Vec
# Crear una lista de listas de palabras (tokens) para entrenar Word2Vec
sentences = df_separaciones['tokens'].tolist()

# Entrenar el modelo Word2Vec
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# 3. Obtener embeddings para cada frase
# Función para promediar los vectores de palabras en cada oración
def obtener_embedding(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if vectors:
        return sum(vectors) / len(vectors)  # Promedio de vectores
    else:
        return [0] * model.vector_size  # Si no hay palabras en el vocabulario, retornar ceros
```

```python
# Aplicar la función para obtener los embeddings
df_separaciones['embedding'] = df_separaciones['tokens'].apply(lambda x: obtener_embedding(x, model))

df_separaciones
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Supongamos que ya tienes el modelo Word2Vec y el DataFrame con embeddings

# Función para procesar la query y obtener su embedding
def procesar_query(query, model):
    tokens = word_tokenize(query.lower())
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if vectors:
        return sum(vectors) / len(vectors)  # Promedio de vectores
    else:
        return np.zeros(model.vector_size)  # Vector de ceros si no hay palabras válidas

# Define tu query
query = "Artificial intelligence in education"

# Generar el embedding de la query
query_embedding = procesar_query(query, model)

# Calcular similitud
def calcular_similitud(query_embedding, embeddings):
    embeddings_matrix = np.stack(embeddings)
    return cosine_similarity([query_embedding], embeddings_matrix)[0]

# Obtener embeddings de las frases
embeddings_list = df_separaciones['embedding'].tolist()

# Calcular similitudes
similaridades = calcular_similitud(query_embedding, embeddings_list)

# Encontrar el resultado más similar
indice_mas_similar = np.argmax(similaridades)
resultado_mas_similar = df_separaciones.iloc[indice_mas_similar]

# Imprimir la información solicitada
print("Query:", query)
print("Resultado más similar:", resultado_mas_similar['sentence'])
print("ID del episodio:", resultado_mas_similar['id'])
print("ID individual:", resultado_mas_similar['st_id'])
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Asegúrate de que la columna de embeddings es una lista de numpy arrays
embeddings = np.array(df_separaciones['embedding'].tolist())

# Número de clusters (esto puedes ajustarlo según tus necesidades)
n_clusters = 5

# Aplicar K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_separaciones['cluster'] = kmeans.fit_predict(embeddings)
```

```python
df_separaciones
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Obtener los centroides de cada cluster
centroides = kmeans.cluster_centers_

# Función para encontrar la similitud coseno entre un vector y el centroide de un cluster
def obtener_similitud_con_centroide(embedding, centroide):
    return cosine_similarity([embedding], [centroide])[0][0]

# Crear un diccionario para almacenar las oraciones más representativas de cada cluster
oraciones_topico = {}

# Iterar sobre cada cluster y encontrar las oraciones más cercanas al centroide
for cluster_id in range(n_clusters):
    # Obtener las oraciones en este cluster
    oraciones_cluster = df_separaciones[df_separaciones['cluster'] == cluster_id]
    
    # Obtener los embeddings de estas oraciones
    embeddings_cluster = np.array(oraciones_cluster['embedding'].tolist())
    
    # Calcular la similitud coseno entre cada embedding y el centroide del cluster
    similitudes = [obtener_similitud_con_centroide(embedding, centroides[cluster_id]) for embedding in embeddings_cluster]
    
    # Encontrar el índice de la oración más cercana al centroide
    indice_max_similitud = np.argmax(similitudes)
    
    # Obtener la oración más representativa y asociarla con el cluster
    oracion_representativa = oraciones_cluster.iloc[indice_max_similitud]['sentence']
    
    # Guardar la oración más representativa en el diccionario
    oraciones_topico[cluster_id] = oracion_representativa

# Mostrar las oraciones más representativas de cada cluster (tópico)
for cluster_id, oracion in oraciones_topico.items():
    print(f"Cluster {cluster_id}: {oracion}")
```
